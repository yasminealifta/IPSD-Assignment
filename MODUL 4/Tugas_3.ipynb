# **Identifikasi pada data tersebut apakah terdapat perbedaan karakteristik data train pada praktikum sebelumnya.**

```python
import pandas as pd
```

```python
test = pd.read_csv('/content/test_features.csv')
train = pd.read_csv('/content/train_features.csv')
```

```python
tf = pd.concat([test, train], axis=1)
```

```python
tf
```

```python
tf.info()
```

```python
tf.describe().T
```

```python
print("Statistik deskriptif data test:")
test.describe().T
```

```python
print("\nStatistik deskriptif data train:")
train.describe().T
```

Berdasarkan statistik deskriptif data train dan data test menunjukkan adanya perbedaan karakteristik. Pada jumlah sampel datanya sendiri, sata test memiliki 3818 dan data train 3817 yang artinya data test memiliki sedikit lebih banyak sampel dibanding sampel data train. Perbedaan tersebut berada pada kolom ID yang terdapat pada data test dan tidak terdapat pada data train. Perbedaan yang paling mencolok juga terlihat pada variabel pendapatan, di mana data test memiliki nilai maksimum dan standar deviasi yang jauh lebih tinggi, menunjukkan adanya lebih banyak kasus ekstrem. Pada kategori belanja seperti belanja buah dan belanja daging, yang mana data tes memiliki nilai rata-rata dan maksimum yang lebih tinggi dibanding data train. Terdapat juga perbedaan kecil seperti pada variabel keluhan dapat dilihat bahwa rata-rata di data test lebih besar dibanding rata-rata variabel keluhan di data train. Perbedaan-perbedaan yang ada ini menunjukkan bahwa data test lebih beragam daripada data train.
