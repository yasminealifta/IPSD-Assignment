Berdasarkan info dan statistik deskriptif dari data train dan data test, dapat diidentifikasi beberapa perbedaan:
- Jumlah data: Jumlah data yang dimiliki oleh data test yaitu 3818 yang mana sedikit lebih banyak dibanding data train yaitu 3817.
- Variabel ID hanya terdapat pada data test, data train tidak memiliki
- Pada variabel pendapatan pada data test memiliki nilai mean, max, dan standar deviasi yang lebih tinggi daripada data train.
- Pada variabel pembelian_diskon, data test memiliki nilai maksimum nya 17 yang mana lebih rendah dari nilai maksimum data train yaitu 20.
- Pada variabel terakhir_belanja, data test memiliki nilai maksimum nya 133 yang mana lebih tinggi dari nilai maksimumnya data train yaitu 128
- Pada varibel tahun_kelahiran, data test memiliki range yang lebih panjang yaitu 1829-2009 dibanding range data train yaitu 1899-2000

Dari beberapa variabel diatas sudah dapat dilihat bahwa terdapat perbedaan karakteristik antara data train dengan data test.
